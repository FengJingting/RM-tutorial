```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
# 1 创建一个空白的图像
img = np.zeros((512,512,3), np.uint8)
# 2 绘制图形
# cv.line(img,start,end,color,thickness)
cv.line(img,(0,0),(511,511),(255,0,0),5)
# cv.rectangle(img,leftupper,rightdown,color,thickness)
cv.rectangle(img,(384,0),(510,128),(0,255,0),3)
# cv.circle(img,centerpoint, r, color, thickness)
cv.circle(img,(447,63), 63, (0,0,255), -1)
font = cv.FONT_HERSHEY_SIMPLEX
# cv.putText(img,text,station, font, fontsize,color,thickness,cv.LINE_AA)
cv.putText(img,'OpenCV',(10,500), font, 4,(255,255,255),2,cv.LINE_AA)
# 3 图像展示
cv2.imshow('img', img)
cv2.waitKey(1000)
cv2.destroyWindow('img')
```

```python
# 通道拆分
img = cv.imread("./image/dili.jpg")
plt.imshow(img[:,:,::-1]) # plt中展示图片需要进行通道翻转
b,g,r = cv.split(img)
plt.imshow(b,cmap=plt.cm.Reds) # 选择cmap展示的颜色
img = cv.merge((b,g,r))
plt.imshow(img[:,:,::-1])
```

```python
# 图像通道的转变
img = cv.imread("./image/lan.jpg")
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY) # 灰度图
hsv = cv.cvtColor(img,cv.COLOR_BGR2HSV)
cv.imshow('gray',gray)
cv.waitKey(10000)
cv.destroyWindow('gray')
# plt.imshow(gray,cmap=plt.cm.gray) # 灰度图
plt.imshow(hsv)
'''
cv.cvtColor(input_image,flag)
参数：

input_image: 进行颜色空间转换的图像
flag: 转换类型
cv.COLOR_BGR2GRAY : BGR↔Gray
cv.COLOR_BGR2HSV: BGR→HSV

'''
```

```python
# 图像的加法 要求，行列数，通道数相同
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# 1 读取图像
img1 = cv.imread("./image/view.jpg")
img2 = cv.imread("./image/rain.jpg")

# 2 加法操作
img3 = cv.add(img1,img2) # cv中的加法
img4 = img1+img2 # 直接相加（效果不好）
plt.imshow(img3[:,:,::-1])
# 3 图像显示
# fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,8),dpi=100)
# axes[0].imshow(img3[:,:,::-1])
# axes[0].set_title("cv中的加法")
# axes[1].imshow(img4[:,:,::-1])
# axes[1].set_title("直接相加")
```

```python
# 图像混合
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# 1 读取图像
img1 = cv.imread("./image/view.jpg")
img2 = cv.imread("./image/rain.jpg")

# 2 混合图像
img3 = cv.addWeighted(img1,0.3,img2,0.7,0)
plt.imshow(img3[:,:,::-1])
```
